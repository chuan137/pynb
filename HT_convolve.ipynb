```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import scipy.signal as signal
import cv2
import functools
from packages.helpers import grid, show_images
from packages.hough import hough_kernel
import packages.tifffile as tifffile
```

```python
img0 = tifffile.imread('data/input1.tif')
img  = img0[400:800, -400:]

img0 = tifffile.imread('data/input1-contrast.tif')
img = img0[400:800, -400:]
```

```python
_ = show_images(img, img0)
```

```python
kernel = hough_kernel(rmin=14, rmax=32, x=200, y=200)
_ = show_images(kernel[0], kernel[-1], cmap='gray')
```

```python
# convolve image with kernel
res = []
for k in kernel:
    res.append( signal.fftconvolve(img, k, mode='same') )
```

```python
# show the results

buf = []
for i in range(len(res)):
    buf.append(res[i])
    if len(buf) == 4 or i == len(res) - 1:
        _ = show_images(*buf, size=5)
        buf = []
```
