```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
img = cv2.imread('p88.bmp')

imgshape = (100,100,3)
simg = np.zeros(imgshape,dtype='uint8')
pimg = np.zeros(imgshape,dtype='uint8')

for r in range(imgshape[0]):
    for c in range(imgshape[1]):
        simg[r,c] = img[200+r,600+c]

# histogram cuts
lb = 30.5 
hb = lb+1
#hb = 50.5

# clipping histogram
case = 2

if case == 1:
    for r in range(imgshape[0]):
        for c in range(imgshape[1]):
            if simg[r,c,0] < lb:
                pimg[r,c] = [0,0,0]
            elif simg[r,c,0] < hb:
                pimg[r,c] = simg[r,c]
            else:
                pimg[r,c] = [255,255,255]
                #pimg[r,c] = [0,0,0]

if case == 2:
    for r in range(imgshape[0]):
        for c in range(imgshape[1]):
            if simg[r,c,0] < lb:
                pimg[r,c] = [0,0,0]
            elif simg[r,c,0] < hb:
                pimg[r,c] = simg[r,c] + (230 - hb)
            else:
                #pimg[r,c] = [255,255,255]
                pimg[r,c] = [0,0,0]
                
cv2.imwrite('sample.bmp', simg)
cv2.imwrite('processed.bmp', pimg)

plt.figure()
fig, axes = plt.subplots(2,2, figsize=(12,12))

axes[0,0].imshow(simg)
axes[0,1].imshow(pimg)

s = axes[1,1].hist(simg.flatten(),bins=256)
axes[1,1].set_xlim([20,60])

# dilation
kernel = np.ones((3,3),np.uint8)
dilation = cv2.dilate(pimg,kernel,iterations = 4)
axes[1,0].imshow(dilation)
```

```python
gbs = 3

blur = cv2.GaussianBlur(pimg,(gbs,gbs),0)
edges = cv2.Canny(blur, 30, 55)
circles = cv2.HoughCircles(blur[:,:,0],cv2.cv.CV_HOUGH_GRADIENT,1,10,param1=5,param2=18,minRadius=0,maxRadius=0)
#print circles

if circles is not None:
    for i in circles[0,:]:
        if i[2] < 10.5:
            # draw the outer circle
            cv2.circle(blur,(i[0],i[1]),i[2],(0,255,0),2)
            # draw the center of the circle
            cv2.circle(blur,(i[0],i[1]),2,(0,0,255),3)

plt.figure()
fig, axes = plt.subplots(1,3, figsize=(24,8))
axes[0].imshow(pimg)
axes[1].imshow(blur)
axes[2].imshow(edges, cmap='gray')
```
