```python

# helper functions

%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import cv2
import functools
from packages.helpers import grid
import packages.tifffile as tifffile

bwimshow = functools.partial(plt.imshow, vmin = 0, vmax = 255,
                             cmap = plt.get_cmap('gray'))

def hough_cone(rmin=10, rmax=20, x=1024, y=1024, rs=2, rt=4, gw=0, gsig=6, gh=1.5):
    """generate ring pattern
    @input rmin, rmax
        minimum or maximum ring size
    @input x, y
        size of output image
    @input rs
        step in ring size
    @input rt
        ring thickness
    @input gw, gsig, gh
        gaussian width, sigma and height for fuzzy HT
    """
    ring_sizes = range(rmin, rmax+1, rs)
    img = np.zeros((len(ring_sizes), x , y))
    for i, r in enumerate(ring_sizes):
        cv2.circle(img[i], (x/2, y/2), r, color=255, thickness=rt)
    if gw > 0:
        fimg = np.copy(img)
        kern = np.zeros(2*gw+1)
        for p in range(gw+1):
            kern[gw+p] = gh * np.exp(-p**2 / gsig**2)
            kern[gw-p] = kern[gw+p]
        for x, y in grid(*img.shape[1:]):
            fimg[:,x,y] = np.convolve(img[:,x,y], kern, mode='same')
        img = fimg
    return img
    
def Brillouin(img):
    x, y = img.shape
    xh, yh = x/2, y/2
    img[:xh, :yh] = img[xh-1::-1, yh-1::-1]
    img[:xh, yh:] = img[xh-1::-1, :yh-1:-1]
    img[xh:, yh:] = img[:xh-1:-1, :yh-1:-1]
    img[xh:, :yh] = img[:xh-1:-1, yh-1::-1]
    return img
```

```python
plt.figure()
fig, axes = plt.subplots(1,3, figsize=(24,8))

cone = hough_cone(x=64,y=64,gw=2)
cone = np.array([Brillouin(c) for c in cone])

axes[0].imshow(cone[0], cmap='gray')
axes[1].imshow(cone[2], cmap='gray')
axes[2].imshow(cone[4], cmap='gray')
```
