```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from packages.helpers import grid, show_images
import packages.tifffile as tifffile

img0= tifffile.imread('data/input1.tif')
img = tifffile.imread('data/input-detection.tif')
img = img * 10**15
```

```python
def explore(img1, img2, x, y, th):
    if x < 0 or x >= img1.shape[0]:
        return False
    if y < 0 or y >= img1.shape[1]:
        return False
    if img2[x][y] == 1:
        return False
    if img1[x][y] < th:
        return False
    img2[x][y] = 1
    explore(img1, img2, x-1, y, th)
    explore(img1, img2, x+1, y, th)
    explore(img1, img2, x, y+1, th)
    explore(img1, img2, x, y-1, th)

def explore2(m, x, y, label=1):
    if label > 0:
        label = -label
    if x < 0 or x >= m.shape[0]:
        return False
    if y < 0 or y >= m.shape[1]:
        return False
    if m[x][y] != 1:
        return False
    m[x][y] = label
    explore2(m, x-1, y, label)
    explore2(m, x+1, y, label)
    explore2(m, x, y+1, label)
    explore2(m, x, y-1, label)
```

```python
m1 = np.copy(img); m1[m1<0] = 0
m2 = np.zeros(img.shape)

n = 0
while n<45:
    for x,y in np.array(np.where(m1==m1.max())).transpose():
        th = 0.01*m1[x,y]
        explore(m1, m2, x, y, th)
        m1[np.where(m2==1)] = 0   
    n = n + 1
pixels = np.where(m2==1)

lab = 0
for x0, y0 in grid(*m2.shape):
    if m2[x0, y0] == 1:
        lab = lab - 1
        explore2(m2, x0, y0, lab)
 
```

```python
# Fisrt image is the input image for HT
# Seconde one is the result from contrast adjustment then HT then blob detection; it used as input in this file
# Third one is the output, namely detected rings
show_images(img0, img, -1*m2, size=6);
```
